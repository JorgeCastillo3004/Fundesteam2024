```python
parcial = df[['Rank', 'Artist', 'Track Name', 'Artist_popularity', 'Streams', 'tempo', 'danceability',
             'energy']]
parcial
```

```python
rank = parcial[parcial.Artist_popularity > 90]
rank
```

```python
rank.Artist.unique()
```

```python
taylor_switf = rank[rank.Artist =='Taylor Swift']
taylor_switf
```

```python
taylor_switf.reset_index(drop=True)
```

```python
x = taylor_switf.energy
y = taylor_switf.Rank
plt.scatter(x,y, color='red')
plt.show()
```

```python
df_2 = taylor_switf[['Rank', 'Artist_popularity','Streams','tempo', 'danceability', 'energy']] 
df_2
cor = df_2.corr(method = 'pearson')
cor
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo con valores faltantes
data = {'A': [1, 2, None, 4, 5],
        'B': [None, 4, 6, None, 10],
        'C': [5, None, 3, 2, 1]}
df = pd.DataFrame(data)

# Rellenar los valores faltantes con un valor específico
filled_df = df.fillna(0)  # Rellenar los NaN con el valor 0

print("DataFrame original:")
print(df)

print("\nDataFrame con valores faltantes rellenados:")
filled_df
```

```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'A': [1, 2, 3],
        'B': ['a', 'b', 'c']}
df = pd.DataFrame(data)

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    print("Índice de fila:", index)
    print("Contenido de la fila:")
    print(row.A)
    print(row.B)
    print()  # Imprimir una línea en blanco para separar las filas
```

```python
import pandas as pd

# Creamos un DataFrame de ejemplo
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [2, 4, 6, 8, 10],
    'C': [5, 4, 3, 2, 1]
}
df = pd.DataFrame(data)
df
```

```python
# df.Volume.plot(kind='line', figsize = (10, 6), color = 'blue', linestyle='-')
df.plot(kind='line', figsize=(10, 6))
plt.title('Verificaion de correlacion')
plt.xlabel('Variable 1')
plt.ylabel("Variable 2")
plt.show()
```

```python
correlation_matrix = df.corr()

print(correlation_matrix)

correlation_matrix = df.corr()

# Obtener el valor de correlación entre las variables A y B
correlation_value = correlation_matrix.loc['A', 'B']

print("Correlación entre las variables A y B:", correlation_value)
```
